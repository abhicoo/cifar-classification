```python
# Run some setup code for this notebook.
import random
import numpy as np
from data_utils import load_CIFAR10
import matplotlib.pyplot as plt

# This is a bit of magic to make matplotlib figures appear inline in the notebook
# rather than in a new window.
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

# Some more magic so that the notebook will reload external python modules;
# see http://stackoverflow.com/questions/1907993/autoreload-of-modules-in-ipython
%load_ext autoreload
%autoreload 2
```

```python
cifar10_dir = 'cifar-10-batches-py'
X_train, y_train, X_test, y_test = load_CIFAR10(cifar10_dir)

print("Training Set Data Shape is {0}".format(X_train.shape))
print("Training Set Label Shape is {0}".format(y_train.shape))
print("Test Set Data Shape is {0}".format(X_test.shape))
print("Test Set Label Shape is {0}".format(y_test.shape))
```

```python
classes = ['plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
num_classes = len(classes)
samples_per_class = 7

for y, cls in enumerate(classes):
    idxs = np.flatnonzero(y_train == y)
    idxs = np.random.choice(idxs, samples_per_class, replace=False)
    for i, idx in enumerate(idxs):
        X = X_train[idx]
        plt_idx = i * num_classes + y + 1
        plt.subplot(samples_per_class, num_classes, plt_idx)
        plt.imshow(X.astype('uint8'))
        plt.axis('off')
        if i == 0:
            plt.title(cls)
plt.show()
```

```python
num_training = 5000
mask = range(num_training)
X_train = X_train[mask, :]
y_train = y_train[mask]

num_test = 500
mask = range(num_test)
X_test = X_test[mask, :]
y_test = y_test[mask]
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], -1))
X_test = np.reshape(X_test, (X_test.shape[0], -1))

print("Training Set Data Shape is {0}".format(X_train.shape))
print("Test Set Data Shape is {0}".format(X_test.shape))
```

```python
class KNearestNeighbour(object):
    
    def __init__(self):
        pass
    
    def train(self, X, y):
        self.X_train = X
        self.y_train = y
        
    def predict(self, X, k=1, num_loops=0):
        if num_loops == 0:
          dists = self.compute_distance_no_loop(X)
        elif num_loops == 1:
          dists = self.compute_distance_one_loop(X)
        elif num_loops == 2:
          dists = self.compute_distance_two_loops(X)
        else:
          raise ValueError('Invalid value %d for num_loops' % num_loops)
        
        return self.predict_labels(dists, k=k)
        
    def compute_distance_two_loops(self, X):
        num_test = X.shape[0]
        num_train = self.X_train.shape[0]
        dist = np.zeros((num_test, num_train))
        for i in range(num_test):
            for j in range(num_train):
                dist[i, j] = np.sqrt(np.sum((X[i] - self.X_train[j])**2))
        return dist
    
    def compute_distance_one_loop(self, X):
        num_test = X.shape[0]
        num_train = self.X_train.shape[0]
        dist = np.zeros((num_test, num_train))
        for i in range(num_test):
            dist[i] = np.sqrt(np.sum((X[i] - self.X_train)**2, axis=1))
        return dist
    
    def compute_distance_no_loop(self, X):
        dist = np.sqrt((X**2).sum(axis = 1, keepdims=True) + (self.X_train**2).sum(axis=1) - 2 * X.dot(self.X_train.T))
        return dist
        
    def predict_labels(self, dists, k):
        num_test = dists.shape[0]
        y_pred = np.zeros(num_test)
        k_nearest_labels = self.y_train[(np.argsort(dists, axis = 1)[:, :k])]
        for i in range(num_test):
            y_pred[i] = np.argmax(np.bincount(k_nearest_labels[i]))
        return y_pred
            
```

```python
classifier = KNearestNeighbour()
classifier.train(X_train, y_train)
```

```python
dists = classifier.compute_distance_two_loops(X_test)
print(dists.shape)
```

```python
plt.imshow(dists, interpolation='none')
plt.show()
```

```python
def print_accuracy(dists, k):
    y_test_pred = classifier.predict_labels(dists, k)
    num_test = X_test.shape[0]
    num_correct = np.sum(y_test_pred == y_test)
    accuracy = float(num_correct) / num_test
    print("Got {0} / {1} correct => accuracy: {2}".format(num_correct, num_test, accuracy))
```

```python
print_accuracy(dists, 1)
print_accuracy(dists, 5)
```

```python
dists_one = classifier.compute_distance_one_loop(X_test)
print(dists_one.shape)
```

```python
difference = np.linalg.norm(dists - dists_one, ord='fro')
print("Difference was: {0}".format(difference))
if difference < 0.001:
  print('Good! The distance matrices are the same')
else:
  print('Uh-oh! The distance matrices are different')
```

```python
print_accuracy(dists_one, 1)
print_accuracy(dists_one, 5)
```

```python
def time_function(f, *args):
  import time
  tic = time.time()
  f(*args)
  toc = time.time()
  return toc - tic

two_loop_time = time_function(classifier.compute_distance_two_loops, X_test)
print("Two loop version took {0} seconds".format(two_loop_time))

one_loop_time = time_function(classifier.compute_distance_one_loop, X_test)
print("One loop version took {0} seconds".format(one_loop_time))

no_loop_time = time_function(classifier.compute_distance_no_loop, X_test)
print("No loop version took {0} seconds".format(no_loop_time))
```

```python
num_folds = 5
k_choices = [1, 3, 5, 8, 10, 12, 15, 20, 50, 100]

X_train_folds = np.split(X_train, num_folds)
y_train_folds = np.split(y_train, num_folds)
```

```python
k_to_accuracies = {}
```

```python
for k in k_choices:
    accuracies = []
    for fold in range(num_folds):
        X_training = np.concatenate(X_train_folds[:fold] + X_train_folds[fold+1:])
        y_training = np.concatenate(y_train_folds[:fold] + y_train_folds[fold+1:])
        X_validation = X_train_folds[fold]
        y_validation = y_train_folds[fold]
        classifier = KNearestNeighbour()
        classifier.train(X_training, y_training)
        y_test_pred = classifier.predict(X_validation, k)
        accuracy = np.mean(y_test_pred == y_validation)
        accuracies.append(accuracy)
    k_to_accuracies[k] = accuracies
        
```

```python
print(k_to_accuracies)
```

```python
for k in sorted(k_to_accuracies):
    for accuracy in k_to_accuracies[k]:
        print("k = {0}, accuracy = {1}".format(k, accuracy))
```

```python
for k in k_choices:
  accuracies = k_to_accuracies[k]
  plt.scatter([k] * len(accuracies), accuracies)

# plot the trend line with error bars that correspond to standard deviation
accuracies_mean = np.array([np.mean(v) for k,v in sorted(k_to_accuracies.items())])
accuracies_std = np.array([np.std(v) for k,v in sorted(k_to_accuracies.items())])
plt.errorbar(k_choices, accuracies_mean, yerr=accuracies_std)
plt.title('Cross-validation on k')
plt.xlabel('k')
plt.ylabel('Cross-validation accuracy')
plt.show()
```

```python
# Based on the cross-validation results above, choose the best value for k,   
# retrain the classifier using all the training data, and test it on the test
# data. You should be able to get above 28% accuracy on the test data.
best_k = 10

classifier = KNearestNeighbour()
classifier.train(X_train, y_train)
y_test_pred = classifier.predict(X_test, k=best_k)


num_correct = np.sum(y_test_pred == y_test)
accuracy = float(num_correct) / num_test
print("Got {0} / {1} correct => accuracy: {2}".format(num_correct, num_test, accuracy))
```
